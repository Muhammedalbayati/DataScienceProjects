```python
data=data.rename(columns={"blue":"bluetooth",
                         "fc":"fc_megapixel",
                         "pc":"pc_megapixel",
                         "m_dep":"m_depth"})
```

```python
data.head()
```

```python
dupes=data.duplicated()
dupes.head()
```

```python
#numbers of duplicates rows
sum(dupes)
```

```python
#To drop duplicates
data=data.drop_duplicates()
```

```python
# To find the count of nulls, also you can use data.isna().sum()
data.isnull().sum()
```

```python
# fill null values with 0
data['clock_speed']=data['clock_speed'].fillna(0)
data.isnull().sum()
```

```python
#FYI: null values also a unique values
len(data['pc_megapixel'].unique())
```

```python
data['pc_megapixel']=data['pc_megapixel'].fillna(method='backfill')
len(data['pc_megapixel'].unique())
```

```python
data.isna().sum()
```

```python
#fill nulls values with median of a column.
data['m_depth']=data['m_depth'].fillna(data['m_depth'].median())
data.isnull().sum()
```

```python
#drop all records with null values
data=data.dropna()
```

```python
#save the data set to CSV file. index=False to not include the index of data frame to the CSV file.
data.to_csv('mobile_data_cleaned.csv',index=False)
#data=pd.read_csv('./mobile_data_cleaned.csv')
```

```python
# show CSV files in the current dirctory
import os
files = [f for f in os.listdir('.') if os.path.isfile(f)]
for f in files:
    extension = os.path.splitext(f)[1]
    if extension=='.csv':
        print(f)
```
