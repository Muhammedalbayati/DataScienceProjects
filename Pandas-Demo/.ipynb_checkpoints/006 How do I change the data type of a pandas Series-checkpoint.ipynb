```python
# good example when numbers are stored as string, or date stored as string
import pandas as pd
drinks=pd.read_csv('http://bit.ly/drinksbycountry')
drinks.head()
```

```python
drinks.dtypes
```

```python
#change column type from int to float
drinks['beer_servings']=drinks.beer_servings.astype(float)
drinks.dtypes
```

```python
# change data type before reading the file
drinks=pd.read_csv('http://bit.ly/drinksbycountry',dtype={'beer_servings':float})
drinks.dtypes
```

```python
orders=pd.read_table('http://bit.ly/chiporders')
orders.head()
```

```python
orders.dtypes
```

```python
# we try to remove the "$" from the orginal numbers and calculate the mean for this column, this will give us an Error, because the item_price is an object, so we need to convert the data type
orders.item_price.str.replace('$','',regex=False).mean()
```

```python
# now we need to convert the data type
orders.item_price.str.replace('$','',regex=False).astype(float).mean()
```

```python
# Other example: here we need to convert the True False to 0 and 1
orders.item_name.str.contains('Chicken').head()
```
