```python
import pandas as pd 
ufo=pd.read_csv('http://bit.ly/uforeports')
ufo.head()
```

```python
# everytime you run this code you will get random rows
ufo.sample(n=3)
```

```python
ufo.sample(n=3, random_state=1) # random_state this will keep the same sample
```

```python
# we want to split the data for two sets one 75% of the rows and the othre is 25%
train=ufo.sample(frac=0.75, random_state=1) # this will randomly chose 75% of your data
train
```

```python
test=ufo.loc[~ufo.index.isin(train.index),:] # this will give us data that not overlab with train dataframe
test
```
