```python
import pandas as pd
# df=pd.read_excel("music.xlsx")
sales_df=pd.read_excel("SalesData3.xlsx",'Sales')#sheet name is case sensitive
companies_df=pd.read_excel("SalesData3.xlsx",'Companies')
managers_df=pd.read_excel("SalesData3.xlsx",'Managers')
```

```python
sales_df.head(3)
```

```python
companies_df.columns
```

```python
#df1.merge(df2, left_on='lkey', right_on='rkey')
df3=pd.merge(sales_df,companies_df[['Name','industry', 'Sector']],left_on='Company Name',right_on='Name').drop(columns= ['Name'])
df3.head(3)
```

```python
managers_df.head(3)
```

```python
df4=pd.merge(df3,managers_df,on='Region')
df4.head(3)
```

```python
#table = pd.pivot_table(df, values='D', index=['A', 'B'],columns=['C'], aggfunc=np.sum)
piv1=pd.pivot_table(df4,index=["Sector","Manager"],values="Sale Amount",aggfunc="sum")
piv1.head(10)
```

```python
#first way to create pivot tabe
piv2=pd.pivot_table(df4,index="Sector",columns='Manager',values="Sale Amount",aggfunc="sum")
piv2
```

```python
#second way to create pivot tabe
ser=df4.groupby(['Sector','Manager'])[['Sale Amount']].sum()
ser.unstack()
```

```python
piv3=pd.pivot_table(df4,index="Sector",values="Sale Amount",aggfunc="sum",margins=True,margins_name="Total")
piv3
```
